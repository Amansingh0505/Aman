```python
import pandas as pd
import datetime as dt
# import matplotlib.pyplot as plt

df = pd.read_csv("D:/PYTHON/WS CUBE TECH/Datasets/archive/dairy_dataset.csv")
df.columns
```

```python
# 1. does number of cows in a farm has any dependency on area aquired. 
gb = df.groupby("Farm Size").agg({"Number of Cows": "sum", "Total Land Area (acres)": "sum"} )
gb
```

```python
x = gb.corr()
x
```

```python
df["Production Date"] = pd.to_datetime(df["Production Date"])
df["Expiration Date"] = pd.to_datetime(df["Expiration Date"])

```

```python
# 2. what is the average expiry days of each product. 
df["Days Count"] = df["Expiration Date"]-df["Production Date"]
df['Days Count'] = df['Days Count'].dt.days
Expiry_days_gb = df.groupby("Product Name").agg(Average_Expiry_Days =("Days Count", "mean"))

Expiry_days_gb
```

```python
# 3. mention the name of the products with max and min shelf life
gb1 = df.groupby("Product Name").agg(Maximum_self_life = ("Shelf Life (days)", "max"))
gb2 = df.groupby("Product Name").agg(Minimum_self_life = ("Shelf Life (days)", "min"))

x = pd.merge(left = gb1, right = gb2, on= "Product Name", how= "inner")
x

```

```python
# 4. summarize the data on the basis of brands and land acquired by them,cows and the revenue generated by them. 
Brand_summary = df.groupby("Brand").agg(Total_Land_Acuired =("Total Land Area (acres)","sum"), Number_Of_Cows = ("Number of Cows","sum"),Total_Revenue= ("Approx. Total Revenue(INR)","sum"))
Brand_summary
```

```python
# 5. What is the average shelf life for different storage conditions. 
gb = df.groupby("Storage Condition").agg(Average_Self_Life_In_Days = ("Days Count", "mean"))
gb
```
